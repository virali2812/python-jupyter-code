# Problem 6 and 9

## Which month saw the (#6) largest and (#9) smallest total revenue, and what was the value?

`Total Revenue = Product Price * Product Quantity * (1 + State Tax Rate)`

---
# Naive Approach using `.apply()`

```python
products = pd.read_csv(url_prefix + "products" + url_suffix, index_col="ProductId")
products.head()
```

```python
sales = [ pd.read_csv(f"{url_prefix}sales-{i:0>2}.csv", dtype={"CustomerId":str}) for i in range(1, 13)]
sales[1].head()
```

```python
mo = sales[9].copy()

mo.loc[:, "Price"] = mo["ProductId"].apply(lambda pid: products.loc[pid, "Price"])
mo.loc[:, "Cost"] = mo["ProductId"].apply(lambda pid: products.loc[pid, "Cost"])

mo.loc[:, "Tax"] = mo["CustomerId"] \
.apply(lambda cid: customers.loc[cid, "State"]) \
.apply(lambda abbr: us_states[abbr]) \
.apply(lambda st: tax_rates.loc[st, "Rate"])

mo.head()
```

```python
s = sum(mo["Quantity"] * mo["Price"] * (1 + mo["Tax"]))
print(f"$ {s:,.2f}")
#print(f"{months[m[1]]} had the most revenue with $ {m[0]:,.2f}.")
```

# Solutions 6 and 9

```python
def get_monthly_revenue(mo):
    df = mo.copy()
    df.drop(["Date", "InvoiceNumber"], axis=1, inplace=True)
    df = pd.merge(df, products, left_on="ProductId", right_index=True, how="outer")
    df.drop("ProductId", axis=1, inplace=True)
    df = pd.merge(df, customers["State"], left_on="CustomerId", right_index=True, how="outer")
    df.drop("CustomerId", axis=1, inplace=True)
    df.rename(columns={"State": "Abbreviation"}, inplace=True)
    df = pd.merge(df, states_df, left_on="Abbreviation", right_index=True, how="outer")
    df.drop("Abbreviation", axis=1, inplace=True)
    df = pd.merge(df, tax_rates["Rate"], left_on="State", right_index=True)
    df.rename(columns={"Rate": "Tax"}, inplace=True)
    df.drop("State", axis=1, inplace=True)
    df.reset_index(inplace=True)
    df.drop("index", axis=1, inplace=True)
    return (df["Price"] * df["Quantity"] * (1 + df["Tax"])).sum()

# df = get_monthly_revenue(sales[9])
# df

mo_revenues = pd.DataFrame([get_monthly_revenue(mo) for mo in sales], columns=["Revenue"], index=months)
mo_revenues["RevenueReadable"] = mo_revenues["Revenue"].apply(lambda x: f"$ {x:,.2f}")
p6 = [mo_revenues["Revenue"][i] for i in range(0,12)]
mo_revenues.sort_values("Revenue", inplace=True)
mo_revenues
#print(months[0], mo_revenues["RevenueReadable"][0])
```

# Problems 7 and 10

## Which month saw the (#7) largest and (#10) smallest gross profit, and what was the value?

`Gross Profit = (Product Price - Product Cost) * Product Quantity`

# Solutions 7 and 10

```python
def get_monthly_gross_profit(mo):
    df = mo.copy()
    df.drop(["Date", "InvoiceNumber", "CustomerId"], axis=1, inplace=True)
    df = pd.merge(df, products, left_on="ProductId", right_index=True, how="outer")
    return ((df["Price"] - df["Cost"]) * df["Quantity"]).sum()

mo_gross_profits = pd.DataFrame([get_monthly_gross_profit(mo) for mo in sales], columns=["GrossProfit"], index=months)
mo_gross_profits["GrossProfitReadable"] = mo_gross_profits["GrossProfit"].apply(lambda x: f"$ {x:,.2f}")
p7 = [mo_gross_profits["GrossProfit"][i] for i in range(0,12)]
mo_gross_profits.sort_values("GrossProfit", inplace=True)
mo_gross_profits
```

---
# Problem 8 and 11

## Which month saw the (#8) largest and (#11) smallest net profit, and what was the value?

`Net Profit = Gross Profit - Business Expenses`

```python
expenses = pd.read_csv(url_prefix + "expenses" + url_suffix)
expenses["Month"] = expenses["Month"].apply(lambda x: months[x-1])
expenses.set_index("Month", inplace=True)
expenses
```

# Solutions 8 and 11

```python
mo_net_profits = mo_gross_profits.copy()
mo_net_profits["NetProfit"] = mo_net_profits["GrossProfit"] - expenses["Expenses"]
mo_net_profits.drop(["GrossProfit", "GrossProfitReadable"], axis=1, inplace=True)
mo_net_profits["NetProfitReadable"] = mo_net_profits["NetProfit"].apply(lambda x: f"$ {x:,.2f}")
p8 = [mo_net_profits["NetProfit"][i] for i in range(0,12)]
mo_net_profits.sort_values("NetProfit", inplace=True)
mo_net_profits
```

---
# Problem 6-11 in one table

```python
df = pd.merge(mo_revenues, mo_gross_profits, left_index=True, right_index=True)
df = pd.merge(df, mo_net_profits, left_index=True, right_index=True)
df.sort_values("NetProfit", inplace=True)
df.drop(["Revenue","GrossProfit", "NetProfit"], axis=1, inplace=True)
df
```

# Problem 12 and 15

## What product generated the (#12) most and (#15) least revenue over the entire year, and what was the amount?

`Price * Quantity * ( 1 + State Sales Tax Rate )`

```python
all_sales = pd.concat(sales)
all_sales.head
```

---
# Solutions 12 and 15

```python
df = all_sales.copy()
df.drop(["Date", "InvoiceNumber"], axis=1, inplace=True)
df = pd.merge(df, products, left_on="ProductId", right_index=True, how="outer")
# df.drop("ProductId", axis=1, inplace=True)
df = pd.merge(df, customers["State"], left_on="CustomerId", right_index=True)
df.drop("CustomerId", axis=1, inplace=True)
df.rename(columns={"State": "Abbreviation"}, inplace=True)
df = pd.merge(df, states_df, left_on="Abbreviation", right_index=True)
df.drop("Abbreviation", axis=1, inplace=True)
df = pd.merge(df, tax_rates["Rate"], left_on="State", right_index=True)
df.rename(columns={"Rate": "Tax"}, inplace=True)
df.drop("State", axis=1, inplace=True)
df.reset_index(inplace=True)
df.drop("index", axis=1, inplace=True)
df["Revenue"] = (df["Price"] * df["Quantity"] * (1 + df["Tax"]))

df
```

```python
product_revenues = df.groupby(["ProductId"])["Revenue"].sum().reset_index()
product_revenues.set_index("ProductId", inplace=True)
product_revenues.sort_values("ProductId")
p9 = product_revenues["Revenue"].apply(lambda x: f"{x:}")
# print(p9)
product_revenues.sort_values("Revenue", inplace=True)
product_revenues["RevenueReadable"] = product_revenues["Revenue"].apply(lambda x: f"$ {x:,.2f}")
product_revenues
```

# Problem 13 and 16

## What product sold the (#13) best and (#16) worst in terms of quantity (Total Quantity Sold), and what was the number?

# Solutions 13 and 16

```python
df = all_sales.copy()
product_qualitities = df.groupby(["ProductId"])["Quantity"].sum().reset_index()
product_qualitities.set_index("ProductId", inplace=True)
product_qualitities["QuantityReadable"] = product_qualitities["Quantity"].apply(lambda x: f" {x:,}")
product_qualitities.sort_values("ProductId")
product_qualitities.sort_values("Quantity", inplace=True)
product_qualitities
```

---
# Problem 14 and 17

## What product was the (#14) most and the (#17) least profitable over the entire year (Gross Profit), and what was the amount?
`Gross Profit = Price - Cost * Quantity`

# Solutions 14 and 17

```python
df = all_sales.copy()
df.drop(["InvoiceNumber", "CustomerId", "Date"], axis=1, inplace=True)
df = pd.merge(df, products, left_on="ProductId", right_index=True)
df["GrossProfit"] = (df["Price"] - df["Cost"]) * df["Quantity"]

df
```

```python
product_gross_profits = df.groupby(["ProductId"])["GrossProfit"].sum().reset_index()
product_gross_profits.set_index("ProductId", inplace=True)
product_gross_profits["GrossProfitReadable"] = product_gross_profits["GrossProfit"].apply(lambda x: f"$ {x:,.2f}")
product_qualitities.sort_values("ProductId")
p10 = product_gross_profits["GrossProfit"].apply(lambda x: f"{x:}")
product_gross_profits.sort_values("GrossProfit", inplace=True)
product_gross_profits
```

---
# Problem 12-17 in one table

```python
df = pd.merge(product_revenues, product_qualitities, left_index=True, right_index=True)
df = pd.merge(df, product_gross_profits, left_index=True, right_index=True)
df.sort_values("GrossProfit", inplace=True)
df.drop(["Revenue", "Quantity", "GrossProfit"], axis=1, inplace=True)

df
```
